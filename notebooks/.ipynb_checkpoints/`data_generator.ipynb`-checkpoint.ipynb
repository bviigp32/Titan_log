```python
import json
import time
import random
from datetime import datetime
from faker import Faker
from minio import Minio
import io

# --- ì„¤ì • (Configuration) ---
MINIO_ENDPOINT = "titan-minio:9000" # Docker ë‚´ë¶€ í†µì‹  ì£¼ì†Œ
ACCESS_KEY = "minioadmin"
SECRET_KEY = "minioadmin"
BUCKET_NAME = "bronze"

fake = Faker()

# 1. MinIO ì—°ê²°
client = Minio(
    MINIO_ENDPOINT,
    access_key=ACCESS_KEY,
    secret_key=SECRET_KEY,
    secure=False
)

# 2. ê°€ì§œ ë°ì´í„° ìƒì„± í•¨ìˆ˜
def generate_log():
    # ì´ë²¤íŠ¸ ìœ í˜• ë¹„ìœ¨: ì¡°íšŒ(70%), ì¥ë°”êµ¬ë‹ˆ(20%), êµ¬ë§¤(10%)
    event_type = random.choices(['view', 'cart', 'purchase'], weights=[0.7, 0.2, 0.1])[0]
    
    log = {
        "event_id": fake.uuid4(),
        "user_id": random.randint(1000, 1100), # ìœ ì € 100ëª… (1000~1100ë²ˆ)
        "item_id": random.randint(1, 50),      # ìƒí’ˆ 50ê°œ
        "event_type": event_type,
        "timestamp": datetime.now().isoformat(),
        "device_os": random.choice(['Android', 'iOS', 'Web']),
        "ip_address": fake.ipv4()
    }
    return log

# 3. ë°ì´í„° ìƒì„± ë° ì—…ë¡œë“œ ì‹¤í–‰
def run_generator(batch_count=10):
    print(f"ë°ì´í„° ìƒì„± ì‹œì‘... (ì´ {batch_count}ê°œì˜ íŒŒì¼ ìƒì„± ì˜ˆì •)")
    
    for i in range(batch_count):
        logs = []
        # íŒŒì¼ í•˜ë‚˜ë‹¹ ë¡œê·¸ 1000ê°œì”© ê½‰ê½‰ ì±„ìš°ê¸°
        for _ in range(1000):
            logs.append(generate_log())
            
        # JSON ë¬¸ìì—´ë¡œ ë³€í™˜
        data_bytes = json.dumps(logs).encode('utf-8')
        data_stream = io.BytesIO(data_bytes)
        
        # íŒŒì¼ëª… ìƒì„± (ì˜ˆ: log_20260125_123456.json)
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        file_name = f"log_date={datetime.now().strftime('%Y-%m-%d')}/log_{timestamp}_{i}.json"
        
        # MinIO ì—…ë¡œë“œ
        client.put_object(
            BUCKET_NAME,
            file_name,
            data_stream,
            length=len(data_bytes),
            content_type='application/json'
        )
        print(f"Uploaded: {file_name} (Size: {len(data_bytes)/1024:.2f} KB)")
        
        time.sleep(1) # 1ì´ˆ ëŒ€ê¸° (ë„ˆë¬´ ë¹ ë¥´ë©´ ì¬ë¯¸ ì—†ìœ¼ë‹ˆê¹Œ)

    print("ëª¨ë“  ë°ì´í„° ìƒì„± ì™„ë£Œ!")

# --- ì‹¤í–‰ ---
# íŒŒì¼ 5ê°œë§Œ ë§Œë“¤ì–´ë´…ì‹œë‹¤. (ì´ 5,000ê±´ì˜ ë¡œê·¸)
if __name__ == "__main__":
    # ë²„í‚·ì´ ì—†ìœ¼ë©´ ìƒì„±
    if not client.bucket_exists(BUCKET_NAME):
        client.make_bucket(BUCKET_NAME)
    
    run_generator(batch_count=5)
```
